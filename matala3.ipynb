```python
import pandas as pd
import numpy as np
import requests
import json
from google.colab import drive
```

```python
from google.colab import drive
drive.mount('/content/drive')
path="/content/drive/MyDrive/data/"
filename ='dests.txt'
```

```python
with open(path+ filename, 'r') as file:
    destinations = file.read().splitlines()
```

```python
df = pd.DataFrame(columns=['Target', 'Distance_km', 'Duration', 'Longitude', 'Latitude'])
for dest in destinations:
    response = requests.get(f'https://maps.googleapis.com/maps/api/distancematrix/json?units=metric&origins=Tel+Aviv&destinations={dest}&key=AIzaSyDajtUhPgYtbEyNgSTXdZQfiY5HUeihzLQ')
    data = response.json()

    duration = data['rows'][0]['elements'][0]['duration']['text']
    distance_km = data['rows'][0]['elements'][0]['distance']['value'] / 1000


    response = requests.get(f'https://maps.googleapis.com/maps/api/geocode/json?address={dest}&key=AIzaSyDajtUhPgYtbEyNgSTXdZQfiY5HUeihzLQ')
    data = response.json()

    longitude = data['results'][0]['geometry']['location']['lng']
    latitude = data['results'][0]['geometry']['location']['lat']

    df = df.append({'Target': dest, 'Distance_km': distance_km, 'Duration': duration, 'Longitude': longitude, 'Latitude': latitude}, ignore_index=True)
```
